Use scipy library to create a randam uniform distribution data (100000 numbers from 0 to 10)

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# random seed
np.random.seed(42)

# range and size of the dist
start = 0
end = 10
size = 100000

# random uniform distribution
data = stats.uniform.rvs(loc=start, scale=end-start, size=size)

# histogram of the data
plt.hist(data, bins=50, density=True, alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Freq')
plt.title('Random Uniform Dist')
plt.show()
```

cdf (cumulative distribution function) gives you the area under the distribution's density curve to the left of a certain value on the x axis. Calculate cdf with x axis from 0 to 2.5.

```python
import numpy as np
from scipy.stats import uniform

# params for the uniform dist
start = 0
end = 10

# uniform dist object
dist = uniform(loc=start, scale=end-start)

# cdf for the range from 0 to 2.5
x = np.linspace(start, 2.5, 100)
cdf_values = dist.cdf(x)
print(f'\nCDF = {cdf_values[-1]}\n\n')
# plot the cdf
import matplotlib.pyplot as plt
plt.plot(x, cdf_values)
plt.xlabel('X-axis')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Uniform Dist')
plt.grid(True)
plt.show()
```

ppf() is the inverse of cdf(): it returns the x axis cutoff value (quantile) associated with a given probability. Caculate the cutoff value for propabality with 40%.

```python
import numpy as np
from scipy.stats import uniform

# params of the uniform distribution
start = 0
end = 10

# calc the cutoff value for prob 0.4
prob = 0.4
cutoff_value = uniform.ppf(prob, loc=start, scale=end-start)

print("Cutoff value for probability 0.4:", cutoff_value)
```

Demonstrate for uniform distribution, x values within the range will have the same probability desity.  x values outside the range have a probability density of 0. The output is similar as the following.

```python
import numpy as np
from scipy.stats import uniform

# Parameters for the uniform distribution
start = 0
end = 10

# Create the uniform distribution object
dist = uniform(loc=start, scale=end-start)

# Define the x values to evaluate
x_values = [-1, 2, 5, 8, 11]

# Calculate and print the probability density for each x value
for x in x_values:
    density = dist.pdf(x)
    print("Density at x value", x)
    print(density)
    print()
```

## The Normal Distribution


Use scipy's norm to have normal distribution (see the graph below) .  The normal distribution has default vaues of  mean of 0 and standard deviation of 1. Investigate what percentage of data generated by a default normal distribution is below -1. What percentage of data generated is above 1 and what percentage lies between -1 and 1. Print the three percentages. The output is similar as the following.

```python
import numpy as np
from scipy.stats import norm

# normal distribution object with mean=0 and std dev=1
dist = norm(loc=0, scale=1)

# calc the percentage of data below -1
p_below_minus_1 = dist.cdf(-1)

# calc the percentage of data above 1
p_above_1 = 1 - dist.cdf(1)

# calc the percentage of data between -1 and 1
p_between_minus_1_and_1 = dist.cdf(1) - dist.cdf(-1)

# print percents
print(p_below_minus_1, p_above_1, p_between_minus_1_and_1)
```

## The Binomial Distribution

We will use binomial distribution to model coin flips. The scipy name for the binomial is binom. Here we set the number of flips per trial to 10,  the number of trials to 10000, and success probability to 0.5.

Create a random binomial distribution data "fair_coin_flips" in the code line below.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

# params for the binomial dist
p = 0.5
n_flips = 10
n_trials = 10000

# rand binomial dist data with the binom.rvs function
fair_flips = binom.rvs(n=n_flips, p=p, size=n_trials)



# Codes are provided here.
# Print out the counts for 0 heads, 1 head, ...10 heads of each trial with 10000 trials.
print( pd.crosstab(index="counts", columns= fair_flips))
# Draw the histogram.
pd.DataFrame(fair_flips).hist(range=(-0.5,10.5), bins=11)
```

The histogram above shows us that a binomial distribution with a 50% probability of success is roughly symmetric, with the most likely outcomes lying at the center.

Continue from above binomial distribution(success rate=0.8), give the codes to get the probability of achieving 5 heads or less with 10 flips.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

# Params for the binomial dist
p = 0.8
n_flips = 10
n_trials = 10000

# new param is number of successful outcomes
# we're looking for 5 or less
max_heads = 5

# rand binomial dist data with the binom.rvs function
fair_coin_flips = binom.rvs(n=n_flips, p=p, size=n_trials)

# using the binom.cdf function to determine the
# probability of achieving 5 heads or less
chance_5_heads_or_less = binom.cdf(k=max_heads, n=n_flips, p=p)

print(chance_5_heads_or_less)
```

Continue from above binomial distribution(success rate=0.8), give the codes to get the probability of achieving 9 heads or more with 10 flips.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

# Params for the binomial dist
p = 0.8
n_flips = 10
n_trials = 10000

# new param is number of successful outcomes
# we're looking for 9 or more so 8 or below is
# how we set the variable since survival func
# is 1 - cdf(x)
# this is setup to say we want everything except
# everything up to 8
max_heads = 8

# rand binomial dist data with the binom.rvs function
fair_coin_flips = binom.rvs(n=n_flips, p=p, size=n_trials)

# using the binom.sf function to determine the
# probability of achieving 9 heads or more
# subtracting 1 because we count from zero
chance_9_heads_or_more = binom.sf(k=max_heads, n=n_flips, p=p)

print(chance_9_heads_or_more)
```

For continuous probability density functions, you use pdf() to check the probability density at a given x value. For discrete distributions like the binomial, use stats.distribution.pmf() (probability mass function) to check the mass (proportion of observations) at given number of successes k.
Write codes to give the probability density at 5 heads with success rate of 0.5 of 10 flips.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

# params for the binomial dist
p = 0.5
n_flips = 10

# probability mass function at 5 heads
num_heads = 5
pmf_5_heads = binom.pmf(k=num_heads, n=n_flips, p=p)

print(pmf_5_heads)
```
